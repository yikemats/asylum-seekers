```python
import requests
from bs4 import BeautifulSoup

import pandas as pd
```

```python
html = open("table_data.html", "r")
```

```python
index = html.read()
```

```python
soup = BeautifulSoup(index, 'lxml')
```

```python
soup
```

```python
tables = soup.select("table")
hotel_table = tables[0]
hotel_table.text[:100]
```

```python
header_row = hotel_table.select("th")
[header.text.strip() for header in header_row]
```

```python
hotel_rows = hotel_table.select("tr")[1:]
hotel_rows
```

```python
hotel_contents = []
for hotel_row in hotel_rows:
    hotel_cells = []
    for cell in hotel_row.select("td"):
        hotel_cells.append(cell.text.strip())
    hotel_contents.append(hotel_cells)
print(hotel_contents)
```

```python
header_contents = [header.text.strip() for header in header_row]
hotel_df = pd.DataFrame(hotel_contents, columns = header_contents)
hotel_df
```

```python
hotel_df['Address'] = hotel_df['Address'].astype(str) + ', NY'
print(hotel_df.head())
```

```python
hotel_df.to_csv('hotel_list.csv')
```
